### Example using the function quadratic f(x) = x^2

```python
#Defining the function
def f(x):
    return x ** 2
```

```python
#Creating a set of random values of x
#linspace python function generates a set of random values from -10 until 10, 400 different values
x = np.linspace(-10,10,400)
```

```python
# calc the y values to each x value
y =f(x)
```

```python
y
```

```python
#Creating a graph
plt.figure(figsize = (6 , 4))
plt.plot(x, y, label = 'f(x) = x^2')

#Adding the title e axis label
plt.title('Quadratic Function Graph')
plt.xlabel('x')
plt.ylabel('f(x)')

# Adding the legend
plt.legend()

#Showing the graph
plt.grid(True)
plt.show()
```

### Sine(sin), Cosine(cos) and Tangent
- Sine (sin): A trigonometric function that gives the y-coordinate of a point on the unit circle as the angle changes. It describes the vertical position in wave-like patterns.

- Cosine (cos): A trigonometric function that gives the x-coordinate of a point on the unit circle as the angle changes. It describes the horizontal position in wave-like patterns

- SINE (sin) 

```python
import math
#Calc the sine of 30 degrees
degrees_angle = 30

#Convert angle from degree to radian
degrees_radian = math.radians(degrees_angle)

#Calc sine from angle, that now is in radians
sine = math.sin(degrees_radian)

print("Sine of 30°: ", sine)
```

- COSINE(cos) 

```python
degrees_angle = 45
radians_angle = math.radians(degrees_angle)
cos = math.cos(radians_angle)
print('Cosine of 45°: ', cos)
```

- TANGENT(tan)

```python
degrees_angle = 60
radians_angle = math.radians(degrees_angle)
tan = math.tan(radians_angle)
print('Tangent of 45°: ', tan)
```

## Logarith (log) Transformation

```python
a, b, c = 5 , 3, 2
# Logarith of the product
product_log = math.log(a * b)

# Logarith of the quotient
quotient_log = math.log(a / b)

#Logarith of the power
power_log = math.log(a**c)

#Property of the product : log(a*b) = log(a) + log(b)
print('Logarith of the product: ', product_log, 'Property: ', math.log(a) + math.log(b))

#Property of the Quotient: log(a/b) = log(a) - log(b)
print('Logarith of the quotient: ', quotient_log, 'Property: ', math.log(a) - math.log(b))

#Property of the Power: log(a**c) = c * log(a) 
print('Logarith of the power: ', power_log, 'Property: ', c * math.log(a))
```

- Example of Log transformation in Data Science 

```python
import numpy as np
import matplotlib.pyplot as plt

#generation data : house prices
np.random.seed(0)
house_prices = np.random.lognormal(mean = 13, sigma = 0.6, size = 1000)

#appling the logarithmic transformation
log_house_prices = np.log(house_prices)

#creating graphs for comparison
plt.figure(figsize = (8, 4))

#Original data graph
plt.subplot(1, 2, 1)
plt.hist(house_prices, bins = 30, color ='blue', alpha = 0.7)
plt.title('Home Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

#Transformed data graph
plt.subplot(1, 2, 2)
plt.hist(log_house_prices, bins = 30, color ='green', alpha = 0.7)
plt.title('Home Price Logarithmic Distribution')
plt.xlabel('Log of Price')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
