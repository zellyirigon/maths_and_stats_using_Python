```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

### 1 - Create a graph for the function y = 2x + 3 for values of x in the interval from 5 to 10.

```python
# Linspace creates a set of random values for x
# np.linspace(start, interval, end)
x = np.linspace(5,10, 100)
y = 2 * x + 3
```

```python
# Plot
plt.figure(figsize = (10,6))
plt.plot(x, y, label = 'y = 2x+3')
plt.title('Function Graph')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)
plt.show()
```

### 2 - Draw the graph of the function y = x^2 - 4x + 4. Create a sample data set using np.linspace().

```python
x = np.linspace(-2,6,100)
y = x**2 - 4 * x + 4

#Plot
plt.figure(figsize = (10,6))
plt.plot(x, y, label = 'y = x^2 - 4x + 4')
plt.title('Function Graph 2')
plt.xlabel('X')
plt.ylabel('Y')

plt.legend()

plt.grid(True)
plt.show()
         
```

### 3 - Create a graph of the function y = sin(x) for values of x in the interval from 0 to 2π.

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

# Plot
plt.figure(figsize = (10,6))
plt.plot(x, y, label = 'y = sin(x)')
plt.title('Sine Graph')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
